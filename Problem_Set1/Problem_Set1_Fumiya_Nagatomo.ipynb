12. Use BeautifulSoup to scrape the data from https://en.wikipedia.org/wiki/Chico,_CA contained in the two table columns outlined in purple below (i.e., the “Race/Ethncity” column and the “2020 Population” column):
Remember to use the Selector Gadget extension to get the CSS selectors. Concatenate the columns into one dataframe. Paste in a screenshot of the dataframe. (Don’t worry about removing the \n characters, you can just leave them there.)


```python
#Answer
import requests
import bs4

url = 'https://en.wikipedia.org/wiki/Chico,_CA'
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/129.0.0.0 Safari/537.36",}
page = requests.get(url, headers = headers)
soup = bs4.BeautifulSoup(page.content, "html.parser")
```

```python
page_info = soup.select('.wikitable caption+ tbody td:nth-child(4) , .mw-heading3+ .wikitable td:nth-child(1)')
```

```python
page_info
```

```python
page_text = [row.get_text() for row in page_info]
```

```python
page_df = pd.DataFrame({
    'Race/Ethnicity':page_text[0::2],
    'Pop2020':page_text[1::2]
})
```
